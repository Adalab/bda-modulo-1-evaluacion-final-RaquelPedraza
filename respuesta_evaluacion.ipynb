```python
import re

class TiendaOnline:
    def __init__(self):

        #  dentro del método constructor definimos los Atributos como 'variables de instancia (POO)' para que sean fácilmente accesibles desde todos los métodos de la clase:

        self.inventario = [] 
                            
        self.clientes = {} 

        self.ventas_totales = 0.0

    # 1-.   
    def agregar_producto(self, nombre, precio, cantidad):
        # Se agrega un nuevo producto o se actualiza su cantidad si ya existe

        # Iteramos a través del inventario
        for producto in self.inventario:
            
            # Comparamos si está el nombre proporcionado entre los nombres de inventario 
            if producto['nombre'] == nombre:
            # Si el producto ya existe, actualizamos su cantidad
                producto['cantidad'] += cantidad
                print(f"Se han agregado {cantidad} unidades del producto '{nombre}'.")
                return self.inventario
            
            # Si el producto no está en el inventario, lo agregamos a éste
        else:
            self.inventario.append({'nombre': nombre, 'precio': precio, 'cantidad': cantidad})
            print (f"El producto '{nombre}' ha sido agregado al inventario con éxito.")
            return self.inventario

    # 2-.     
    def ver_inventario(self): 
        # Muestra el inventario de productos con sus detalles

        # Utilizamos un bucle for para recorrer el inventario
        for producto in self.inventario:
            # Extrae los detalles de cada producto
            nombre = producto['nombre'].capitalize() 
            precio = producto['precio']
            cantidad = producto['cantidad']

            # Imprimimos estos detalles 
            print(f'Nombre: {nombre}, Precio: ${precio}, Cantidad: {precio} ')

    # 3-. 
    def buscar_producto(self, nombre):
        # (Buscamos un producto en el inventario por nombre)

        # Iteramos el inventario con un bucle for
        for producto in self.inventario:
            # Comparamos si está el nombre proporcionado entre los nombres de inventario 
            if producto['nombre'] == nombre:
                # Si encuentra el producto, imprime sus detalles
                print(f'Nombre: {nombre.capitalize()}, Precio: ${producto["precio"]}, Cantidad: {producto["cantidad"]} ')
            # Si el producto no está en el inventario, mostramos un mensaje 
            else:
                print (f"El producto '{nombre}' no existe en inventario.") 

    
    # 4-.
    def actualizar_stock(self, nombre, cantidad): 
 
        # Iteramos el inventario con un bucle for
        for producto in self.inventario:
            # Comparamos si está el nombre proporcionado entre los nombres de inventario 
            if producto['nombre'] == nombre:
                # Si encuentra el producto, pide al usuario que actualice la cantidad:
                print(f'Nombre: {nombre.capitalize()}, Cantidad: {producto["cantidad"]} ')
                producto['cantidad'] += cantidad
                print(f'Stock de {nombre} actualizado con éxito. {nombre.capitalize()} = {producto["cantidad"]} unidades.')

            # Si el producto no está en el inventario, mostramos un mensaje 
            else:
                print (f'El producto "{nombre}" no existe en inventario.') 
    
    # 5-.
    def eliminar_producto(self, nombre): 
        
        # Iteramos el inventario con un bucle for
        for producto in self.inventario:
            # Comparamos si está el nombre proporcionado entre los nombres de inventario 
            if producto['nombre'] == nombre:
                self.inventario.remove(producto)
                print(f'El producto "{nombre}" ha sido eliminado correctamente del inventario.')
                print(f'El inventario queda: {self.inventario}.')
            
            # Si el producto no está en el inventario, mostramos un mensaje 
            
        print (f'El producto "{nombre}" no existe en inventario.') 
        # return None

    # 6-.
    def calcular_valor_inventario(self): 

        # Inicializamos la variable donde vamos a guardar el valor total del inventario
        valor_total = 0

        #Iteramos sobre el inventario
        for producto in self.inventario:
            # Calculamos el valor de cada producto multiplicando el precio de éste por la cantidad en stock
            valor_producto = producto['precio'] * producto['cantidad']
            # Sumamos el valor de cada producto para llegar al valor total
            valor_total += valor_producto
             
        # Devolvemos el valor total deseado
        print(f'El valor total de los prouctos en almacén es de ${valor_total}.')

    
    # 7-. 
    def buscar_producto_regex(self, patron): 

        # Creamos una lista para almacenar los productos que coincidan con el patrón
        coincidencias_patron = []

        #Iteramos sobre el inventario
        for producto in self.inventario:
            # Utilizamos re.match
            if re.match(patron, producto['nombre']):
                # Agregamos la coincidencia a la lista antes creada 
                coincidencias_patron.append(producto)

        # Imprimimos los detalles de los productos si hubo coincidencias
        if coincidencias_patron:
                print(f'Las coincidencias con el patrón regex "{patron}", son las siguientes: {coincidencias_patron}.')
                for producto in coincidencias_patron:
                    print(f"Nombre: {producto['nombre']}, Precio: ${producto['precio']}, Cantidad: {producto['cantidad']}")
        else:
            print(f'No se encontraron productos que coincidan con el patrón "{patron}".')

        # Devolvemos la lista de productos que coinciden con el patrón
        return coincidencias_patron

    # 8-.
    def realizar_compra(self):   

        print('Bienvenido a TiendaOnline. Este es nuestro catálogo de productos: \n')
        # Iteramos sobre el inventario y lo mostramos:
        for producto in self.inventario:
            print(f"| Nombre: {producto['nombre'].capitalize()}, Precio: ${producto['precio']}, Cantidad disponible: {producto['cantidad']} |")

    
        carrito = [] #?? no dicc
        coste_total = 0.0


        # Con un bucle while permitimos al cliente realizar múltiples compras
        pedido = ''
        while pedido != 'fin':
            # Solicitamos al cliente ingresar el nombre del producto que desa comprar
            pedido = input("Seleccione el nombre del producto que desea comprar (o 'fin' para terminar la compra): ")
            if pedido.lower() == 'fin':
                break

            cantidad = int(input("¿Cuántas unidades del producto desea?"))
                
            for producto in self.inventario:
                if producto['nombre'].lower() == pedido:
                    if cantidad <= producto['cantidad']:
                        # Registramos los productos seleccionados en un carrito
                        carrito.append({'nombre': producto['nombre'], 'precio': producto['precio'], 'cantidad': cantidad})
                        # Actualizamos el stock de inventario
                        producto['cantidad'] -= cantidad
                        # Calculamos el costo total de la compra
                        coste_total += cantidad * producto['precio']
                        print(f"\n Producto '{producto['nombre']}' x {cantidad} uds añadido al carrito.")
                        break
        
        else:
            print(f"\n El producto {pedido} no está disponible o no existe en el inventario.")
     
        print(f"\n El costo total de la compra es: ${coste_total:.2f}")
        
        return carrito
    
    
    # 9-. 
    def procesar_pago(self): 
    
        try:

            # Solicitamos al cliente ingresar la cantidad total y la cantidad de pago
            cant_total = float(input("Ingrese la cantidad total de la compra"))
            cant_pago = float(input("Escriba la cantidad que va a usar para pagar"))
            
            # Si la cantidad introducida es suficiente
            if cant_pago >= cant_total:
                # Calculamos el cambio
                cambio = cant_pago - cant_total
                # Y mostramos un mensaje de pago exitoso
                print(f'Su cuenta está pagada. El cambio es de ${cambio:.2f}. ¡Gracias por su compra!')
            else:
                # Mostramos un mensaje de error si la cantidad de pago es insuficiente
                print("Error: La cantidad de pago es insuficiente para cubrir el total de la compra.")

        except ValueError:

             # Manejamos la excepción si ocurre un error al convertir la entrada a números de punto flotante
            print("Error: Por favor, ingrese cantidades válidas como números.")
        
    # 10-. 
    def agregar_cliente(self, nombre, email): 
        
        # Agregamos un nuevo cliente al diccionario clientes (nombre = clave, email y compras = valores)
        self.clientes[nombre] = {'email': email, 'compras': []}
        print(f'El cliente "{nombre}" ha sido correctamente agregado a la lista de clientes.')
        return nombre
    
    # 11-.
    def ver_clientes(self): 

        print("Lista de clientes registrados:")
        # Iteramos con un for por la base de datos clientes
        for nombre, datos in self.clientes.items():
            # Imprimimos los detalles de cada cliente (nombre y correo electrónico).
            print(f"Cliente: {nombre}, e-mail: {datos['email']}")

    # 12-. 
    def registrar_compra(self, nombre_cliente, carrito): 
        
        # Buscamos al cliente en el diccionario de clientes
        if nombre_cliente in self.clientes:
            # Calculamos el total de la compra
            total_compra = sum(producto['precio'] * producto['cantidad'] for producto in carrito.values())
            # Actualizamos las ventas totales de la tienda
            self.ventas_totales += total_compra
            # Registramos la compra, incluyendo los productos y el total
            self.clientes[nombre_cliente]['compras'].append({'productos': carrito, 'total': total_compra})
            print(f"Compra registrada para el cliente '{nombre_cliente}' por un total de ${total_compra:.2f}.")
        
        else: 
            # Mostramos mensaje para avisar de que es un cliente no registrado
            print(f"No se puede anotar la compra por que el cliente '{nombre_cliente}' no está registrado.") 
    
    # 13-. 
    def ver_compras_cliente(self, nombre_cliente): 
        # (Muestra el historial de compras de un cliente)
         
        # Buscamos al cliente en el diccionario de clientes
        if nombre_cliente in self.clientes:
            historial_compras = self.clientes[nombre_cliente]['compras']
            if historial_compras:
                print(f"Historial de compras para el cliente '{nombre_cliente}':")
                # Iteramos sobre cada compra en el historial
                for compra in historial_compras:
                    # Imprimimos un encabezado para cada compra
                    print("Compra:")
                    # Iteramos sobre cada producto en la compra
                    for producto, detalles in compra['productos'].items():
                        print(f"Producto: {producto}, Precio: ${detalles['precio']}, Cantidad: {detalles['cantidad']}")
                print(f"Total de la compra: ${compra['total']:.2f}")
                # Imprimimos un separador entre cada compra
                print("-" * 30)
            else:
                print(f"El cliente '{nombre_cliente}' no tiene historial de compras.")
        else:
            print(f"No se encontró al cliente '{nombre_cliente}' en el registro.")

    # 14-. 
    def calcular_ventas_totales(self):
        # Sumamos los totales de todas las compras realizadas 
        total_ventas = sum(compra['total'] for cliente in self.clientes.values() for compra in cliente['compras'])
        # Mostramos las ventas totales en la tienda
        print(f"Las ventas totales de la tienda son: ${total_ventas:.2f}")
        

 
```
